#### Elements of a plot

In Matplotlib, the figure (an instance of the class plt.Figure) can be thought of as a single container that contains all the objects representing axes, graphics, text, and labels.  

The axes (an instance of the class plt.Axes) is what we see above: a bounding box with ticks and labels, which will eventually contain the plot elements that make up our visualization

### Subplots

- multiple axes (plots) in a single figure

```python
# Within a figure - create two subplots (one row and two columns)

plt.subplot(1,2,1) #(number of rows, number of columns, number subplot)
plt.text(0.5, 0.5, str((1,2,1)), ha = 'center', fontsize = 14)

#plt.plot()

plt.subplot(1,2,2) 
plt.text(0.5, 0.5, str((1,2,2)), ha = 'center', fontsize = 14)

#plt.scatter()
```

```python
x = [x * 0.25 for x in range(-25,26)]
y1 = [math.sin(x) for x in x]
y2 = [math.cos(x) for x in x]
y3 = [math.sin(x) * -1 for x in x]
y4 = [math.cos(x) * -1 for x in x]
```

```python
# 2 columns and 2 rows
pi = math.pi
plt.figure(figsize = (10, 8))

plt.suptitle('Sin and Cos functions', fontsize = 16, color = 'w')

plt.subplot(2,2,1)
plt.plot( x, y1, ls = '-', marker = '*', color = 'red')
plt.title('$sin(x)$', color = 'w')
plt.grid(True, ls = '--')

# X tick values to show up as gradiants multiples of five
plt.xticks([-2 * pi , -1 * pi, 0, pi, 2 * pi]), ['$-2\pi$', '$-\pi$', '0', '$\pi$', '$2\pi$'] # Changes the scale for x axis

plt.subplot(2,2,2)
plt.plot( x, y2, ls = '-', marker = 'v', color = 'green')
plt.title('$cos(x)$', color = 'w')
plt.grid(True, ls = '--')
plt.axhline(0, color = 'k', linewidth = 0.5) # Makes a horizontal line at y = 0 

plt.subplot(2,2,3)
plt.plot( x, y3, ls = '-', marker = '*', color = 'blue')
plt.title('$-sin(x)$', color = 'w')
plt.grid(True, ls = '--')

plt.subplot(2,2,4)
plt.plot( x, y4, ls = '-', marker = 'o', color = 'k')
plt.title('$-cos(x)$', color = 'w')
plt.grid(True, ls = '--')
plt.axhline(0, color = 'k', linewidth = 0.5)
plt.fill_between(x, y4, color = 'blue') # Fills the area in the graph with a color.

#Add padding to avoid overlap of subplots 
plt.tight_layout()

plt.show()


```

```python
plt.figure(figsize = (10, 6))
x1 = [random.gauss(0, 0.8) for _ in range(1000)] #(mean 0, sigma 0.8)
x2 = [random.gauss(-2, 1) for _ in range(1000)]
x3 = [random.gauss(3, 1.5) for _ in range(1000)]
x4 = [random.gauss(7, 0.6) for _ in range(1000)]

kwargs = dict(alpha = 0.3, bins = 30, density = True)

plt.hist(x1, **kwargs) # use ** to pass on kwargs to each histogram.
plt.hist(x2, **kwargs)
plt.hist(x3, **kwargs)
plt.hist(x4, **kwargs)

plt.show()
```

```python
# use for loop to make subplots
plt.figure(figsize = (10,6)) 

# Make list of list ti integrate
x = [x1, x2, x3, x4]

for i in range(1, 5):
    plt.subplot(2,2,i)
    plt.hist(x[i-1], color = 'g', edgecolor = 'k') # Starts with 1 not 0 so use (i- 1)
```
